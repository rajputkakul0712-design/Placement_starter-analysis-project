## Function 1 — load_placements(filename)

**Skills:** `csv.reader` (Day 17) · `try/except FileNotFoundError` (Day 18)

````python
def load_placements(filename):
    """Read placement_data.csv and return a list of rows."""
    placements = []

    try:
        with open(filename, "r", newline="", encoding="utf-8") as file:
            reader = csv.reader(file)
            next(reader)          # Skip header
            for row in reader:
                placements.append(row)

    except FileNotFoundError:
        print("File not found.")

    return placements

# === TEST 1 ===
placements = load_placements("placement_data.csv")
print(f"Loaded {len(placements)} records")
assert len(placements) >= 245, f"Expected ~250 records, got {len(placements)}"
print(f"First row: {placements[0]}")
print("TEST 1 PASSED")
````

## Function 2 — display_placement(row)

**Output:** `  Aarav Sharma (CSE | CGPA: 8.40) -- TCS, Software Engineer, Bangalore -- Rs.6.50 LPA`  
Show `N/A` for missing cgpa or package.

**Skills:** `try/except ValueError` (Day 18) · f-strings (Day 4)

````python
def display_placement(row):
    """Print one-line placement summary."""

    name = row[1]
    branch = row[2]
    company = row[5]
    role = row[6]

    try:
        cgpa = f"{float(row[3]):.2f}"
    except:
        cgpa = "N/A"

    try:
        package = f"{float(row[4]):.2f}"
    except:
        package = "N/A"

    print(f"{name} ({branch} | CGPA: {cgpa}) -> {company} | {role} | Rs.{package} LPA")
# === TEST 2 ===
print("Sample records:")
for p in placements[:5]:
    display_placement(p)
print("TEST 2 PASSED (check output format above)")
````

## Function 3 — avg_package_by_branch(placements)

Average package (LPA) per branch. Skip non-numeric packages.

**Skills:** `dict.get()` accumulator (Day 12) · `try/except ValueError` (Day 18)

````python
def avg_package_by_branch(placements):
    """Average package by branch."""

    totals = {}
    counts = {}

    for row in placements:
        branch = row[2]

        try:
            package = float(row[4])
        except:
            continue

        if branch not in totals:
            totals[branch] = 0
            counts[branch] = 0

        totals[branch] += package
        counts[branch] += 1

    return {b: totals[b] / counts[b] for b in totals}

# === TEST 3 ===
branch_avgs = avg_package_by_branch(placements)
print(f"Branches: {len(branch_avgs)}")
assert len(branch_avgs) >= 4
for b, avg in sorted(branch_avgs.items(), key=lambda x: x[1], reverse=True):
    print(f"  {b}: Rs.{avg:.2f} LPA")
print("TEST 3 PASSED")
````

## Function 4 — top_companies(placements, n=5)

N companies that hired the most students.

**Skills:** `dict.get()` counter (Day 12) · sort with lambda (Day 11)

````python
def top_companies(placements, n=5):
    """Top companies by number of hires."""

    company_counts = {}

    for row in placements:
        company = row[5]

        if company not in company_counts:
            company_counts[company] = 0

        company_counts[company] += 1

    return sorted(company_counts.items(),
                  key=lambda x: x[1],
                  reverse=True)[:n]
# === TEST 4 ===
companies = top_companies(placements, 5)
print("Top 5 companies:")
for co, count in companies:
    print(f"  {co}: {count} students")
assert len(companies) == 5
print("TEST 4 PASSED")
````

## Function 5 — highest_packages(placements, n=5)

Top N students by package. Skip non-numeric packages.

**Skills:** build list of (value, row), sort, slice (Days 11–12)

````python
def highest_packages(placements, n=5):
    """Top students by package."""

    earners = []

    for row in placements:
        try:
            package = float(row[4])
            earners.append(row)
        except:
            pass

    earners.sort(key=lambda x: float(x[4]), reverse=True)

    return earners[:n]

# === TEST 5 ===
top5 = highest_packages(placements, 5)
assert len(top5) == 5, f"Expected 5, got {len(top5)}"
print("Top 5 packages:")
for i, p in enumerate(top5, 1):
    print(f"  {i}. {p[1]} ({p[2]}) -- Rs.{p[4]} LPA at {p[5]}")
print("TEST 5 PASSED")
````

## Function 6 — write_html_report(placements, branch_avgs, companies, top5, filename)

**Replace every `???` with the correct Python variable or expression.**  
Do not add or remove any other lines.

**Skills:** knowing your own data — `branch_avgs` is a dict, `companies` is a list of tuples, `top5` is a list of rows.

````python
def write_html_report(placements, branch_avgs, companies, top5, filename):
    """Write a styled HTML campus placement report.
    Structure: header > summary cards > branch table > companies > top packages > footer.

    branch_avgs -- dict {branch: avg_lpa}
    companies   -- list of (company, count) tuples
    top5        -- list of placement rows

    Row index reference:
        row[0]=student_id  row[1]=name  row[2]=branch  row[3]=cgpa
        row[4]=package_lpa  row[5]=company  row[6]=role  row[7]=city
    """

    total = len(placements)
    top_co = companies[0][0] if companies else "N/A"

    with open(filename, "w") as f:

        write_html_header(
            f,
            "Campus Placement Report",
            f"Placement analysis -- {total} students (2020-2024)"
        )

        # ---------- SUMMARY CARDS ----------
        start_cards(f)
        add_card(f, "Students Placed", total, "total records")
        add_card(f, "Top Recruiter", top_co, f"{companies[0][1]} hires", "accent")
        add_card(f, "Highest Package", f"Rs.{top5[0][4]} LPA", top5[0][1], "warm")
        add_card(f, "Branches", len(branch_avgs), "branches tracked", "dark")
        end_cards(f)

        # ---------- BRANCH AVERAGES ----------
        start_section(f, "Average Package by Branch")
        start_table(f, ["Branch", "Avg Package (LPA)"])

        for branch, avg in sorted(branch_avgs.items(), key=lambda x: x[1], reverse=True):
            add_row(f, [branch, f"Rs.{avg:.2f}"])

        end_table(f)
        end_section(f)

        # ---------- TOP HIRING COMPANIES ----------
        start_section(f, "Top Hiring Companies")
        start_table(f, ["Company", "Students Hired"])

        for company, count in companies:
            add_row(f, [company, count])

        end_table(f)
        end_section(f)

        # ---------- HIGHEST PACKAGES ----------
        start_section(f, "Highest Packages")
        start_table(f, ["Name", "Branch", "Package (LPA)", "Company"])

        for p in top5:
            add_row(f, [p[1], p[2], f"Rs.{p[4]}", p[5]])

        end_table(f)
        end_section(f)

        write_html_footer(f, "Kakul Rajput")
````
