````python
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# PRE-BUILT HTML HELPERS  --  run this cell once, do not modify
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
import os, csv, datetime

REPORT_CSS = """<style>
  * { margin:0; padding:0; box-sizing:border-box; }
  body { font-family:'Segoe UI',Calibri,sans-serif; background:#f4f1eb; color:#2d2a24;
         padding:32px 24px; max-width:960px; margin:0 auto; }
  .report-header { background:linear-gradient(135deg,#2d2a24 0%,#3a3630 100%);
    color:#f8f4ed; padding:36px 32px; border-radius:12px; margin-bottom:28px; }
  .report-header h1 { font-size:28px; font-weight:700; margin-bottom:6px; }
  .report-header p  { font-size:14px; color:#c4cebc; }
  .summary-cards { display:grid; grid-template-columns:repeat(auto-fit,minmax(180px,1fr));
    gap:16px; margin-bottom:28px; }
  .card { background:#fff; border-radius:10px; padding:20px;
    box-shadow:0 2px 8px rgba(45,42,36,.08); border-left:4px solid #8e9a7c; }
  .card.accent { border-left-color:#b26b4a; }
  .card.warm   { border-left-color:#d4956f; }
  .card.dark   { border-left-color:#2d2a24; }
  .card .label  { font-size:12px; text-transform:uppercase; letter-spacing:.5px;
                  color:#6e665c; margin-bottom:6px; }
  .card .value  { font-size:28px; font-weight:700; color:#2d2a24; }
  .card .detail { font-size:13px; color:#6e665c; margin-top:4px; }
  section { background:#fff; border-radius:10px; padding:24px; margin-bottom:24px;
            box-shadow:0 2px 8px rgba(45,42,36,.08); }
  section h2 { font-size:18px; font-weight:600; color:#2d2a24; margin-bottom:16px;
               padding-bottom:8px; border-bottom:2px solid #e0d6c7; }
  table { width:100%; border-collapse:collapse; font-size:14px; }
  thead th { background:#2d2a24; color:#f8f4ed; padding:10px 14px;
             text-align:left; font-weight:600; font-size:13px; }
  thead th:first-child { border-radius:6px 0 0 0; }
  thead th:last-child  { border-radius:0 6px 0 0; }
  tbody td { padding:9px 14px; border-bottom:1px solid #e0d6c7; }
  tbody tr:nth-child(even) { background:#faf6f0; }
  tbody tr:hover { background:#ede8dc; }
  .footer { text-align:center; font-size:12px; color:#6e665c;
            margin-top:32px; padding-top:16px; border-top:1px solid #e0d6c7; }
</style>"""


def write_html_header(f, title, subtitle=""):
    """Opens the HTML file and writes the dark header banner."""
    f.write("<!DOCTYPE html>\n<html lang='en'>\n<head>\n")
    f.write(f"<meta charset='UTF-8'>\n<title>{title}</title>\n")
    f.write(REPORT_CSS)
    f.write("</head>\n<body>\n")
    f.write('<div class="report-header">\n')
    f.write(f'  <h1>{title}</h1>\n')
    if subtitle:
        f.write(f'  <p>{subtitle}</p>\n')
    f.write('</div>\n\n')

def write_html_footer(f, author=""):
    """Closes the HTML file and writes the footer line."""
    f.write('<div class="footer">\n')
    f.write(f'  Generated on {datetime.date.today()} using Python')
    if author:
        f.write(f' by {author}')
    f.write('<br>\n  Programming with Python &middot; AKGEC Summer Internship 2026\n')
    f.write('</div>\n</body>\n</html>\n')

# ── Card helpers ────────────────────────────────────────
def start_cards(f):
    """Open the summary card grid. Call before add_card()."""
    f.write('<div class="summary-cards">\n')

def add_card(f, label, value, detail="", style=""):
    """Write one summary card.
    style options: "" (sage border), "accent" (clay), "warm" (orange), "dark"
    """
    cls = f"card {style}".strip()
    f.write(f'  <div class="{cls}">\n')
    f.write(f'    <div class="label">{label}</div>\n')
    f.write(f'    <div class="value">{value}</div>\n')
    if detail:
        f.write(f'    <div class="detail">{detail}</div>\n')
    f.write('  </div>\n')

def end_cards(f):
    """Close the summary card grid. Call after the last add_card()."""
    f.write('</div>\n\n')

# ── Section + table helpers ─────────────────────────────
def start_section(f, heading):
    """Open a white content panel with a heading."""
    f.write('<section>\n')
    f.write(f'  <h2>{heading}</h2>\n')

def end_section(f):
    """Close the content panel."""
    f.write('</section>\n\n')

def start_table(f, columns):
    """Open a table and write the header row.
    columns -- list of column heading strings, e.g. ["Month", "Avg Temp", "Rainy Days"]
    """
    f.write('  <table>\n  <thead><tr>')
    for col in columns:
        f.write(f'<th>{col}</th>')
    f.write('</tr></thead>\n  <tbody>\n')

def add_row(f, values):
    """Write one data row. values -- list of cell contents (strings or numbers)."""
    f.write('  <tr>')
    for v in values:
        f.write(f'<td>{v}</td>')
    f.write('</tr>\n')

def end_table(f):
    """Close the table."""
    f.write('  </tbody>\n  </table>\n')

def add_paragraph(f, text):
    """Write a short paragraph of plain text inside the current section."""
    f.write(f'  <p style="margin:8px 0;font-size:14px;color:#2d2a24;">{text}</p>\n')

print("HTML helpers loaded -- 11 functions ready.")
print("  write_html_header / write_html_footer")
print("  start_cards / add_card / end_cards")
print("  start_section / end_section")
print("  start_table / add_row / end_table / add_paragraph")
````
